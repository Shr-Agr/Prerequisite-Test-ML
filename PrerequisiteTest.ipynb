In multiplication of two matrices the numbers in the rows in the first matrix are multiplied with the corresponding numbers in columns of the second matrix.
If we want to multiply two matrices A(n,k) and B(k,m) then <br>
$P[i,j]=\sum_{l=0}^{k-1} A[i,l]*B[l,j]$<br>
So in multiplication of a n x k matrix with a k x m matrix:<br>
Total Multiplications = $n*m*k$<br>
Total Additions = $n*m*(k-1)$

```python
import numpy as np
import time
n=110
m=102
k=125
Mat_A=np.random.randint(100,size = (n, k))
Mat_B=np.random.randint(100,size = (k, m))
Mat_P=np.zeros((n, m)) #product matrix
#converting the numpy arrays into lists
Mat_A=list(Mat_A)
Mat_B=list(Mat_B)
Mat_P=list(Mat_P)
start_time=time.time()
for i in range(len(Mat_A)):
  for j in range(len(Mat_B[0])):
    for l in range(len(Mat_B)):
      Mat_P[i][j]+=Mat_A[i][l]*Mat_B[l][j]
end_time=time.time()
exec_time=end_time-start_time
print("Time =",exec_time)
print()
print(Mat_P)
```

```python
# multiplication using numpy
Mat_A2=np.asarray(Mat_A)
Mat_B2=np.asarray(Mat_B)
Mat_P2=np.zeros((n, m))
start=time.time()
Mat_P2=np.dot(Mat_A2,Mat_B2)
end=time.time()
time=end-start
print("Time =",time)
print()
print(Mat_P2)
```

The numpy method is much faster since the numpy arrays are densely packed in memory which allows for better cache utilisation. The algorithms in numpy dot are also highly optimised.

The method of finding the median by traversing the array again again and finding the highest, second highest, and so on, elements would take $O(n^2)$ time.

n = number of elements in the array

A single pass would take $O(n)$ complexity and median is the $(n+1)/2th$ largest element in an array (if n is odd) and mean of $n/2th$ and $n/2+1th$ elements if n is even. So the time complexity will be $n*(n+1)/2$ (if n=odd) and $n*n/2$ (if n=even).

So the overall time complexity would be $O(n^2)$

A better approach would be to sort the array in increasing order which would take O(nlogn) time and then just find the median of the array.

```python
import numpy as np
import time
# Implementation of the given method to find the median
def findElement(arr,pos):
  dic={}    #dictionary for storing the max elements found so far
  for i in range(pos):
    max_el=-1
    max_el_ind=-1
    for j in range(len(arr)):
      if dic.get(j)==None:
        max_el=max(max_el,arr[j])
        if max_el==arr[j]:
          max_el_ind=j
    dic[max_el_ind]=max_el
  return max_el
start=time.time()
arr=np.random.randint(1,1000,1000)
size=len(arr)
# print(arr)        ##run this to print the array
if size%2==0:
  num1=findElement(arr,int(size/2))
  num2=findElement(arr,int((size/2)+1))
  print((num1+num2)/2)
else:
  print(findElement(arr,int((size+1)/2)))
end=time.time()
time=end-start
print("Time =",time)
```

```python
import time
#Finding the median by sorting the array
temp_arr=arr
temp_arr.sort()
# print(temp_arr)
start2=time.time()
size=len(arr)
if size%2==0:
  num1=temp_arr[int(size/2 - 1)]
  num2=temp_arr[int(size/2)]
  print((num1+num2)/2)
else:
  print(temp_arr[int((size-1)/2)])
end2=time.time()
time=end2-start2
print("Time =",time)
```

```python
# using numpy
import time
import numpy as np
start=time.time()
temp=list(arr)
print(np.median(temp))
end=time.time()
print(end-start)
```

The numpy.median routine is faster than the other two methods.

```python
def minIndex(arr):
  min=max(arr)+1
  min_ind=len(arr)+1
  for i in range(len(arr)):
    if(min>arr[i]):
      min=arr[i]
      min_ind=i
  return min_ind
print("Min Index:",minIndex(arr))
```
